```python
import pandas as pd
df_train = pd.read_csv("train_merged[961-1120].csv")
df_test = pd.read_csv("test_merged[961-1120].csv")
df_train
```

```python
interaction_count = df_train.groupby(["user_id", "merchant_id"]).size().reset_index(name="interaction_count")

interaction_count
df_train = pd.merge(df_train, interaction_count, on=['user_id', 'merchant_id'])
df_train
```

```python
user_merchant_distinct_categories = df_train.groupby(['user_id', 'merchant_id'])['cat_id'].nunique().reset_index()
user_merchant_distinct_categories.columns = ['user_id', 'merchant_id', 'user_merchant_distinct_categories']
user_merchant_distinct_categories
df_train = pd.merge(df_train, user_merchant_distinct_categories, on=['user_id', 'merchant_id'])
df_train
```

```python
# total number of actions taken by each user
user_actions = df_train.groupby('user_id')['action_type'].count()


# print(user_actions)
# total number of clicks on each item
item_clicks = df_train[df_train['action_type'] == 0].groupby('item_id')['action_type'].count()
print(item_clicks)
# average age of users who purchased from each merchant
merchant_avg_age = df_train[df_train['action_type'] == 2].groupby('merchant_id')['age_range'].mean()
df_train = pd.merge(df_train, user_actions, on='user_id').rename(columns={'action_type_y':'action_by_user'})
df_train = pd.merge(df_train, item_clicks, on='item_id',how='left')
df_train
```
