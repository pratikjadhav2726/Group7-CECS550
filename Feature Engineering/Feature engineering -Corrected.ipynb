```python
import pandas as pd
data_user_log = pd.read_csv("../EDA/df_train_user_log.csv")
data_user_info = pd.read_csv("../EDA/df_train_user_info.csv")

# df_train['time_stamp'] = pd.to_datetime(df_train['time_stamp'], format='%m%d').apply(lambda x: x.replace(year=2022))
# df_test['time_stamp'] = pd.to_datetime(df_test['time_stamp'], format='%m%d').apply(lambda x: x.replace(year=2022))

data_user_info
```

```python
interaction_count = data_user_log.groupby(["user_id", "merchant_id"]).size().reset_index(name="interaction_count")

print(interaction_count)
data_user_info = pd.merge(data_user_info, interaction_count, on=['user_id', 'merchant_id'])
data_user_info
```

```python
user_active_days = data_user_log[['user_id','time_stamp']].drop_duplicates()
user_active_days_info = user_active_days.groupby('user_id').size().reset_index()
user_active_days_info.columns = ['user_id','user_active_days']
user_active_days_info
data_user_info = pd.merge(data_user_info, user_active_days_info, on=['user_id'])
data_user_info
```

```python
cat_item = data_user_log[['cat_id','item_id']].copy()
cat_item.drop_duplicates(['cat_id','item_id'], inplace=True)
cat_item_info = cat_item.groupby('cat_id').size().reset_index()
cat_item_info.columns = ['cat_id','category_item_count']
cat_item_info
```

```python
# Number of distinct categories a user has interacted with for each merchant
distinct_cat_per_user_merchant = data_user_log.groupby(['user_id', 'merchant_id'])['cat_id'].nunique().reset_index()
distinct_cat_per_user_merchant.rename(columns={'cat_id': 'distinct_cat_per_user_merchant'}, inplace=True)
data_user_info = pd.merge(data_user_info, distinct_cat_per_user_merchant, on=['user_id', 'merchant_id'], how='left')
```

```python
data_user_info
```

```python
# Average age of users who interacted with each merchant
age_avg_per_merchant = data_user_info.groupby('merchant_id')['age_range'].mean().reset_index()
age_avg_per_merchant.rename(columns={'age_range': 'age_avg_per_merchant'}, inplace=True)
data_user_info = pd.merge(data_user_info, age_avg_per_merchant, on='merchant_id', how='left')
data_user_info
```

```python
# Number of distinct brands a user has interacted with for each merchant
distinct_brand_per_user_merchant = data_user_log.groupby(['user_id', 'merchant_id'])['brand_id'].nunique().reset_index()
distinct_brand_per_user_merchant.rename(columns={'brand_id': 'distinct_brand_per_user_merchant'}, inplace=True)
data_user_info = pd.merge(data_user_info, distinct_brand_per_user_merchant, on=['user_id', 'merchant_id'], how='left')
data_user_info
```

```python
# Ratio of add-to-cart actions to clicks for each merchant
add_to_cart_to_click_ratio_per_merchant = data_user_log.groupby('merchant_id')['action_type'].apply(lambda x: (x == 1).sum() / (x == 0).sum()).reset_index()
add_to_cart_to_click_ratio_per_merchant.rename(columns={'action_type': 'add_to_cart_to_click_ratio_per_merchant'}, inplace=True)
data_user_info = pd.merge(data_user_info, add_to_cart_to_click_ratio_per_merchant, on='merchant_id', how='left')
data_user_info
```

```python
# Average User Age for each Merchant
avg_user_age_merchant = data_user_info.groupby('merchant_id')['age_range'].mean().reset_index()
avg_user_age_merchant.rename(columns={'age_range':'avg_user_age_merchant'}, inplace=True)
data_user_info = pd.merge(data_user_info, avg_user_age_merchant, on='merchant_id', how='left')
data_user_info
```

```python
# Calculate average user age for each category
avg_user_age_category = data_user_log.groupby('cat_id')['user_id'].apply(lambda x: df_train[df_train['user_id'].isin(x)]['age_range'].mean()).reset_index(name='avg_user_age_category')
avg_user_age_category
# Merge with df_train
# df_train = df_train.merge(avg_user_age_category, on='cat_id', how='left')
```

```python
# Purchase Frequency
purchase_frequency = data_user_log.groupby(['user_id', 'merchant_id'])['action_type'].agg(purchase_count='sum', total_count='count').reset_index()
purchase_frequency['purchase_frequency'] = purchase_frequency['purchase_count'] / purchase_frequency['total_count']
purchase_frequency.drop(['purchase_count', 'total_count'], axis=1, inplace=True)
data_user_info = pd.merge(data_user_info, purchase_frequency, on=['user_id', 'merchant_id'], how='left')
data_user_info
```

```python
# Purchase Ratio
purchase_ratio = data_user_log.groupby(['user_id', 'merchant_id'])['action_type'].agg(purchase_count='sum', total_count='count').reset_index()
purchase_ratio['purchase_ratio'] = purchase_ratio['purchase_count'] / purchase_ratio['total_count']
purchase_ratio.drop(['purchase_count', 'total_count'], axis=1, inplace=True)
data_user_info = pd.merge(data_user_info, purchase_ratio, on=['user_id', 'merchant_id'], how='left')
data_user_info
```

```python


# Purchase Average Time
purchase_average_time = df_train[df_train['action_type'] == 2].sort_values(by=['user_id', 'merchant_id', 'time_stamp'])
purchase_average_time['time_diff'] = purchase_average_time.groupby(['user_id', 'merchant_id'])['time_stamp'].diff()
purchase_average_time = purchase_average_time.groupby(['user_id', 'merchant_id'])['time_diff'].mean().reset_index()
purchase_average_time.rename(columns={'time_diff':'purchase_average_time'}, inplace=True)
purchase_average_time
data_user_info = pd.merge(data_user_info, purchase_average_time, on=['user_id', 'merchant_id'], how='left')
data_user_info
```
