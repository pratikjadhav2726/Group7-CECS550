```python
import pandas as pd
df_train = pd.read_csv("../EDA/train_merged[961-1120].csv")
df_test = pd.read_csv("../EDA/test_merged[961-1120].csv")
# df_train['time_stamp'] = pd.to_datetime(df_train['time_stamp'], format='%m%d').apply(lambda x: x.replace(year=2022))
# df_test['time_stamp'] = pd.to_datetime(df_test['time_stamp'], format='%m%d').apply(lambda x: x.replace(year=2022))

df_train
```

```python
interaction_count = df_train.groupby(["user_id", "merchant_id"]).size().reset_index(name="interaction_count")

interaction_count
df_train = pd.merge(df_train, interaction_count, on=['user_id', 'merchant_id'])
df_train
```

```python
# Number of distinct categories a user has interacted with for each merchant
distinct_cat_per_user_merchant = df_train.groupby(['user_id', 'merchant_id'])['cat_id'].nunique().reset_index()
distinct_cat_per_user_merchant.rename(columns={'cat_id': 'distinct_cat_per_user_merchant'}, inplace=True)
df_train = pd.merge(df_train, distinct_cat_per_user_merchant, on=['user_id', 'merchant_id'], how='left')
```

```python
df_train.head()
```

```python
# Average age of users who interacted with each merchant
age_avg_per_merchant = df_train.groupby('merchant_id')['age_range'].mean().reset_index()
age_avg_per_merchant.rename(columns={'age_range': 'age_avg_per_merchant'}, inplace=True)
df_train = pd.merge(df_train, age_avg_per_merchant, on='merchant_id', how='left')
df_train.head()
```

```python
# Number of distinct brands a user has interacted with for each merchant
distinct_brand_per_user_merchant = df_train.groupby(['user_id', 'merchant_id'])['brand_id'].nunique().reset_index()
distinct_brand_per_user_merchant.rename(columns={'brand_id': 'distinct_brand_per_user_merchant'}, inplace=True)
df_train = pd.merge(df_train, distinct_brand_per_user_merchant, on=['user_id', 'merchant_id'], how='left')
df_train.head()
```

```python
# Ratio of add-to-cart actions to clicks for each merchant
add_to_cart_to_click_ratio_per_merchant = df_train.groupby('merchant_id')['action_type'].apply(lambda x: (x == 1).sum() / (x == 0).sum()).reset_index()
add_to_cart_to_click_ratio_per_merchant.rename(columns={'action_type': 'add_to_cart_to_click_ratio_per_merchant'}, inplace=True)
df_train = pd.merge(df_train, add_to_cart_to_click_ratio_per_merchant, on='merchant_id', how='left')
df_train
```

```python
# Average User Age for each Merchant
avg_user_age_merchant = df_train.groupby('merchant_id')['age_range'].mean().reset_index()
avg_user_age_merchant.rename(columns={'age_range':'avg_user_age_merchant'}, inplace=True)
df_train = pd.merge(df_train, avg_user_age_merchant, on='merchant_id', how='left')
df_train
```

```python
# Repeat Buyer Count for each Merchant
repeat_buyer_count_merchant = df_train.groupby('merchant_id')['label'].sum().reset_index()
repeat_buyer_count_merchant.rename(columns={'label':'repeat_buyer_count_merchant'}, inplace=True)
df_train = pd.merge(df_train, repeat_buyer_count_merchant, on='merchant_id', how='left')
df_train
```

```python
repeat_buyer_count_merchant
```

```python
# Calculate average user age for each category
avg_user_age_category = df_train.groupby('cat_id')['user_id'].apply(lambda x: df_train[df_train['user_id'].isin(x)]['age_range'].mean()).reset_index(name='avg_user_age_category')

# Merge with df_train
df_train = df_train.merge(avg_user_age_category, on='cat_id', how='left')
```

```python
df_train
```

```python
# Purchase Frequency
purchase_frequency = df_train.groupby(['user_id', 'merchant_id'])['action_type'].agg(purchase_count='sum', total_count='count').reset_index()
purchase_frequency['purchase_frequency'] = purchase_frequency['purchase_count'] / purchase_frequency['total_count']
purchase_frequency.drop(['purchase_count', 'total_count'], axis=1, inplace=True)
df_train = pd.merge(df_train, purchase_frequency, on=['user_id', 'merchant_id'], how='left')
df_train
```

```python
# Purchase Ratio
purchase_ratio = df_train.groupby(['user_id', 'merchant_id', 'item_id'])['action_type'].agg(purchase_count='sum', total_count='count').reset_index()
purchase_ratio['purchase_ratio'] = purchase_ratio['purchase_count'] / purchase_ratio['total_count']
purchase_ratio.drop(['purchase_count', 'total_count'], axis=1, inplace=True)
df_train = pd.merge(df_train, purchase_ratio, on=['user_id', 'merchant_id','item_id'], how='left')
df_train
```

```python


# Purchase Average Time
purchase_average_time = df_train[df_train['action_type'] == 2].sort_values(by=['user_id', 'merchant_id', 'time_stamp'])
purchase_average_time['time_diff'] = purchase_average_time.groupby(['user_id', 'merchant_id'])['time_stamp'].diff()
purchase_average_time = purchase_average_time.groupby(['user_id', 'merchant_id'])['time_diff'].mean().reset_index()
purchase_average_time.rename(columns={'time_diff':'purchase_average_time'}, inplace=True)
df_train = pd.merge(df_train, purchase_average_time, on=['user_id', 'merchant_id'], how='left')
df_train
```

```python
df_train.to_csv('feature_Engineered_train.csv', index=False)
```
