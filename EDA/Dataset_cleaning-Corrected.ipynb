```python
#exploring dataset and importing into dataframe
import pandas as pd
df_test_format1 = pd.read_csv("TestFormat1-group7.csv")
print(df_test_format1.shape)
df_test_format1.head(10)
```

```python
#exploring dataset and importing into dataframe

df_train_format1 = pd.read_csv("TrainFromat1-group7.csv")
print(df_train_format1.shape)
df_train_format1.head(10)
```

```python
#exploring dataset and importing into dataframe

df_user_info_format1 = pd.read_csv("user_info_format1.csv")
print(df_user_info_format1.shape)
df_user_info_format1.head(10)
```

```python
#exploring dataset and importing into dataframe

df_user_log_format1 = pd.read_csv("user_log_format1.csv")
print(df_user_log_format1.shape)
df_user_log_format1.head(10)
```

```python
df_test_user_log = pd.merge(df_test_format1,df_user_log_format1,on=['user_id'],how='left')
df_test_user_info = pd.merge(df_test_format1,df_user_info_format1,on=['user_id'],how='left')
print(df_test_user_info.shape)
# df_test.head(10)
df_test_user_info
```

```python
df_train_user_log = pd.merge(df_train_format1,df_user_log_format1,on=['user_id'],how='left')
df_train_user_info = pd.merge(df_train_format1,df_user_info_format1,on=['user_id'],how='left')
print(df_train_user_info.shape)
df_train_user_log
```

```python
#convert float to asint 
df_train_user_info=df_train_user_info.dropna()
df_train_user_info['age_range']=df_train_user_info['age_range'].astype(int)
df_train_user_info['gender']=df_train_user_info['gender'].astype(int)



df_train_user_log=df_train_user_log.dropna()
df_train_user_log['brand_id']=df_train_user_log['brand_id'].astype(int)

```

```python
#convert float to asint Test
df_test_user_info=df_test_user_info.dropna()
df_test_user_info['age_range']=df_test_user_info['age_range'].astype(int)
df_test_user_info['gender']=df_test_user_info['gender'].astype(int)



df_test_user_log=df_test_user_log.dropna()
df_test_user_log['brand_id']=df_test_user_log['brand_id'].astype(int)
```

```python
df_train_user_log.to_csv('df_train_user_log.csv', index=False)
df_test_user_info.to_csv('df_test_user_info.csv', index=False)
df_train_user_info.to_csv('df_train_user_info.csv', index=False)
df_test_user_log.to_csv('df_test_user_log.csv', index=False)
```
